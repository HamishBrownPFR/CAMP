## Bring in MetData to calculate thermal time

```python
#Bring in the APSIM met files
LNmet = pd.read_csv(root+'Data/NPI_Agribio2018/LN.met',sep='\t',skiprows=11)
LNmet.drop(0,inplace=True)
for c in LNmet.columns:
    LNmet.loc[:,c] = pd.to_numeric(LNmet.loc[:,c])
LNmet.loc[:,'date'] = [dt.datetime(LNmet.loc[x,'year'],1,1) + dt.timedelta(float(LNmet.loc[x,'day'])) for x in LNmet.index]
LNmet.set_index('date',inplace=True)
LNmet.drop(['year','day'],axis=1,inplace=True)
LNmet.loc[SowDates['LN']:,'AccumTtSow'] = LNmet.loc[SowDates['LN']:,'mint'].cumsum()

SNmet = pd.read_csv(root+'Data/NPI_Agribio2018/SN.met',sep='\t',skiprows=11)
SNmet.drop(0,inplace=True)
for c in SNmet.columns:
    SNmet.loc[:,c] = pd.to_numeric(SNmet.loc[:,c])
SNmet.loc[:,'date'] = [dt.datetime(SNmet.loc[x,'year'],1,1) + dt.timedelta(float(SNmet.loc[x,'day'])) for x in SNmet.index]
SNmet.set_index('date',inplace=True)
SNmet.drop(['year','day'],axis=1,inplace=True)
SNmet.loc[SowDates['SN']:,'AccumTtSow'] = SNmet.loc[SowDates['SN']:,'mint'].cumsum()


LVmet = pd.read_csv(root+'Data/NPI_Agribio2018/LV.met',sep='\t',skiprows=11)
LVmet.drop(0,inplace=True)
for c in LVmet.columns:
    LVmet.loc[:,c] = pd.to_numeric(LVmet.loc[:,c])
LVmet.loc[:,'date'] = [dt.datetime(LVmet.loc[x,'year'],1,1) + dt.timedelta(float(LVmet.loc[x,'day'])) for x in LVmet.index]
LVmet.set_index('date',inplace=True)
LVmet.drop(['year','day'],axis=1,inplace=True)
LVmet.loc[SowDates['LV']:,'AccumTtSow'] = LVmet.loc[SowDates['LV']:,'mint'].cumsum()

SVmet = pd.read_csv(root+'Data/NPI_Agribio2018/SV.met',sep='\t',skiprows=11)
SVmet.drop(0,inplace=True)
for c in SVmet.columns:
    SVmet.loc[:,c] = pd.to_numeric(SVmet.loc[:,c])
SVmet.loc[:,'date'] = [dt.datetime(SVmet.loc[x,'year'],1,1) + dt.timedelta(float(SVmet.loc[x,'day'])) for x in SVmet.index]
SVmet.set_index('date',inplace=True)
SVmet.drop(['year','day'],axis=1,inplace=True)
SVmet.loc[SowDates['SV']:,'AccumTtSow'] = SVmet.loc[SowDates['SV']:,'mint'].cumsum()

Met = pd.concat([SNmet,LNmet,SVmet,LVmet],keys=['SN','LN','SV','LV'])

#Bring in the datalogger data from in the growth rooms
Temp = pd.read_excel(root+'Data/NPI_Agribio2018/GrowthRoomTemp.xlsx',engine='openpyxl')
Temp.Date = pd.to_datetime(Temp.Date)
Temp.set_index(['Environment','Date'],inplace=True)
CanopyTemp = pd.read_excel(root+'Data/NPI_Agribio2018/GrowthRoomsCanopyTemp.xlsx',header=2,engine='openpyxl')
CanopyTemp.Date = pd.to_datetime(CanopyTemp.Date)
CanopyTemp.set_index(['Environment','Date'],inplace=True)
```

```python
def PlotLNlines(LC,SC,SN,LN,fs):
    MinLN = LC
    PpLN = max(0,SC - LC)
    CvLN = max(0,SN - MinLN - PpLN)
    PvLN = LN - MinLN - CvLN
    
    #MinLN
    plt.plot([.6,3],[MinLN]*2,'--',color='k',lw=2)
    plt.plot([0.6]*2,[0,MinLN],'-',color='k',lw=2)
    plt.text(0.7,LC-2,'$MinLN=$'+"{:.1f}".format(MinLN),fontsize=fs,verticalalignment='center')
    #PpLN
    if PpLN>0.5:
        plt.plot([1.6,2.6],[MinLN+PpLN]*2,'--',color='k',lw=2)
        plt.plot([1.6,1.6],[SC,LC],'-',color='k',lw=2)
        plt.text(1.5,(LC+SC)/2,'$PpLN=$'+"{:.1f}".format(PpLN),fontsize=fs,verticalalignment='center',horizontalalignment='right')
    else:
        plt.text(1.3,(LC+SC)/2,'$PpLN=0$',fontsize=fs,verticalalignment='bottom')
    #CvLN
    if CvLN > 0.5:
        plt.plot([2.6,2.6],[SC,MinLN+PpLN+CvLN],'-',color='k',lw=2)
        plt.text(2.5,(SC+SN)/2,'$CvLN=$'+"{:.1f}".format(CvLN),fontsize=fs, verticalalignment='center',horizontalalignment='right')
        
        plt.plot([2.6,3],[MinLN+PpLN+CvLN,MinLN+CvLN],'--',color='k',lw=2)
        plt.plot([2.6,3],[MinLN+PpLN,MinLN],'--',color='k',lw=2)
        plt.plot([3,3.6],[MinLN + CvLN]*2,'--',color='k',lw=2)
        #plt.plot([3.1,3.1],[LC+CvLN,LC],'-',color='k',lw=2)
    else:
        plt.text(2.1,(SC+SN)/2,'$CvLN=0$',fontsize=fs, verticalalignment='bottom')
    #PvLN
    if ((PvLN > 0.5)or(PvLN<-0.5)):
        plt.plot([3.2,3.6],[LN,LN],'--',color='k',lw=2)
        plt.plot([3.2,3.2],[LC+CvLN,LN],'-',color='k',lw=2)
        plt.text(3.3,LC+CvLN + PvLN/2,'$PvLN=$'+"{:.1f}".format(PvLN),fontsize=fs, verticalalignment='center')
    else:
        plt.text(3.3,LC+CvLN + PvLN/2,'$PvLN=0$',fontsize=fs, verticalalignment='center')
```

```python
def MakeGenotypeBarPlot(Geno,Axis):
    width = 0.4
    ind = np.arange(4) + width
    colors = ['b','b','r','r']
    fills = ['b','w','w','r']
    SubPos = 0
    for Photo,Vern in [('L','V'),('S','V'),('S','N'),('L','N')]:
            try:
                y = FLNData.loc[(Photo,Vern,Geno),'Value']
            except:
                y = []
            try:
                x = np.add(list(range(len(y))),SubPos)
            except:
                x = []
            plt.plot(x,y, 'o', markersize = 12, mec = colors[SubPos],
                     mfc=fills[SubPos], label = Photo + " " + Vern)
            try:
                plt.bar(SubPos,y.median(),0.8,
                        edgecolor=colors[SubPos], color = fills[SubPos],
                        linewidth=3,alpha=0.4)
            except:
                dummy = 'bla bla'
            #plt.xticks([3,9,15,21])
            plt.text(1,24,Geno, fontsize=16);
            plt.ylim(0,28)
            SubPos +=1
    Axis.spines['right'].set_visible(False)
    Axis.spines['top'].set_visible(False)
    plt.tick_params(axis='x',direction='in',labelsize = 12,rotation=30)
    try:
        LC = FLNData.loc[('L','V',cul),'Value'].median()
        SC = FLNData.loc[('S','V',cul),'Value'].median()
        SN = FLNData.loc[('S','N',cul),'Value'].median() 
        LN = FLNData.loc[('L','N',cul),'Value'].median()
        PlotLNlines(LC,SC,SN,LN,8)
        #plt.plot([18,24],[MinLN + VrnLN]*2,'-',color='k',lw=2)
        #plt.text(21,22,str(VxPLN))
    except:
        do = 'nothing'    
    
    if Pos in list(range(1,101,5)):
        plt.tick_params(axis='y', which='both', left=True,right=False, labelleft=True)
        plt.ylabel('FinalLeafNumber',fontsize=16)
    else:
        plt.tick_params(axis='y', which='both', left=True,right=False, labelleft=False)
```
