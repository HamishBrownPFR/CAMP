```python
def convertHStoBP(HS):
    """Calculates how many base phyllochrons have accumulated at the specified HaunStage"""
    return np.interp(HS,
                     [0,2,  7,  30],
                     [0,1.5,6.5,38.7],
                     False)
        
def calcTSHS(FLN,IntFLNvsTSHS):
    """Haun stage timing of terminal spikelet 
       Inverts equation 5 from Brown etal 2013 FLN =  2.85 + 1.1*TSHS and converts it to base phyllochrons.
       Note the intercept differs, was typeo on publication
       Intercept has been made variable for as needs to be lower for some very fast varieties
    Args:
        FLN: The final leaf number
        IntFLNvsTSHS: The intercept of the regression between FLN and TSHS
    Returns:
        Estimation of number of base phyllochrons to terminal spikelet
    """
    return (FLN - IntFLNvsTSHS)/1.1

def deriveVrnParams(c):
    data = pd.Series(dtype=np.float64)
    # Base Phyllochron duration of Emergence Phase
    data['EmergDurat'] = convertHStoBP(CampInputs.loc[c,'TtEmerge']/CampInputs.loc[c,'BasePhyllo'])
    # The minimum accumulated base phyllochrons at which Vern saturation can occur
    data['MinVS'] = convertHStoBP(1.1)
    # Base Phyllochron duration of vernalisation treatment
    data['VernTreatDurat'] = convertHStoBP((CampInputs.loc[c,'VrnTreatDuration']*CampInputs.loc[c,'VrnTreatTemp'])/CampInputs.loc[c,'BasePhyllo'])
    # Accumulated Base Phyllochrons when vernalisation traetment ended 
    data['EndVernTreat'] =  -data['EmergDurat'] + data['VernTreatDurat']
    # The minimum base phyllochron duration from vern saturation to terminal spikelet
    data['MinVS->TS'] = min(convertHStoBP(3),(convertHStoBP(CampInputs.loc[c,'FLN_LV']) - data['MinVS'])/2) #From Lincoln controlled environment data
    # Base Phyllochron duration from terminal spikelet to flag leaf appearance
    data['TS->FL_LV'] = convertHStoBP(CampInputs.loc[c,'FLN_LV']) - data['MinVS'] - data['MinVS->TS']
    # the intercept of the relatinship between Flag Leaf Number and the accumulated base phyllochrons at Terminal Spikelet
    data['IntFLNvsTS'] = min(2.85,data['TS->FL_LV']/1.1)
    # Calculate the accumulated base phyllochrons at terminal spikelet for each treatment
    for pv in ['SN','LV','SV','LN']:
        data['TS_'+pv] = calcTSHS(CampInputs.loc[c,'FLN_'+pv],data['IntFLNvsTS'])
    # Calculate the accumulated base phyllochrons at vernalisation saturation for each treatment
    data['VS_LV'] = data['TS_LV'] - data['MinVS->TS'] 
    data['VS_LN'] = max(data['TS_LN'] - data['MinVS->TS'],data['VS_LV']) #Constrained so not earlier than the LV treatment
    data['VS_SV'] = min(data['VS_LV'],data['TS_SV']-data['MinVS->TS']) # Assume happens at the same time as VS_LV but not sooner than TS and minVS->TS would allow 
    data['VS->TS_SV'] = data['TS_SV'] - data['VS_SV']
    data['VS_SN'] = max(data['TS_SN'] - data['VS->TS_SV'],data['VS_LV']) #Constrained so not earlier than the LV treatment
    # Base Vrn1 delta during vegetative phase.  Assuming base Vrn1 expression starts at sowing and reaches 1 at VS for the SN treatment where no cold or Vrn3 (short Pp) to upregulate 
    data['BaseDVrn1Veg'] = 1/(data['VS_SN']+data['EmergDurat'])
    # Base Vrm1 delta during early reproductive phase.  Assuming Vrn1 expression increases by 1 between VS and TS and proceeds at a base rate where not Vrn3 upregulation (short PP)
    data['BaseDVrn1ER'] = 1/(data['TS_SN']-data['VS_SN'])
    # The maximum Vrn1 delta during the early reproductive phase.  Assuming Vrn1 increases by 1 bewtween VS and TS and proceeds at the maximum rate under long photoperiod
    data['MaxDVrn1ER'] = 1/data['MinVS->TS']
    # Base Phyllochron duration of vernalisation, may be less than treatment duration if treatment goes past VS
    data['VernDurat_SV'] =  min(data['VernTreatDurat'],data['VS_SV']+data['EmergDurat'])
    # The amount of Vrn1 expression at base rate in the SV treatment between sowing and VS 
    data['baseVrn1Veg_SV'] = (data['EmergDurat']+min(data['VS_SV'],data['VS_SN'])) * data['BaseDVrn1Veg']
    # The amount of persistant (methalated) Vrn1 upregulated due to cold at the end of the vernalisation treatment for SV
    data['ColdVrn1_SV'] = 1 - data['baseVrn1Veg_SV']
    # # The base phyllochron duration of effective vernalisation.  Will be the minimum of treatment duration or VS when cold vernalisation response is assumed to stop.
    data['EndVern'] = min(data['EndVernTreat'],data['VS_SV']) 
    # The rate of Vrn1 expression under cold treatment
    data['ColdDVrn1'] = data['ColdVrn1_SV']/data['VernDurat_SV'] 
    # The upregulation of Vrn1 expression above base rate at 0oC
    data['ColdDVrn1_V'] = data['ColdDVrn1']/ np.exp(CampInputs.loc[c,'k'] * CampInputs.loc[c,'VrnTreatTemp']) 
    # The maximum Vrn1 expression at 0oC
    data['MaxColdDVrn1'] = data['ColdDVrn1_V'] + data['BaseDVrn1Veg']
    # The relative increase in delta Vrn1 cuased by Vrn3 expression under long photoperiod
    data['PpVrn1Fact'] = data['MaxDVrn1ER']/data['BaseDVrn1ER']
    # The rate of Vrn1 expression after minVS under long Pp
    data['Vrn1_L'] = data['BaseDVrn1Veg'] * data['PpVrn1Fact']
    # The base phyllochron duration when effective vrn1 is up regulating during the vegetative phase under long photoperiod
    data['Vrn1Durat_L'] = 1/data['Vrn1_L']
    # The accumulated base phyllochron when Vrn2 expression is ended and effective Vrn1 expression starts under long Pp
    data['EndVrn2_L'] = max(0,data['VS_LN'] - data['Vrn1Durat_L'])
    # The Vrn2 Expression that must be matched by Vrn1 before effective Vrn3 expression starts.  
    data['Vrn2_L'] = max(0,data['EndVrn2_L'] * data['BaseDVrn1Veg'])
    return data
```

```python
CultParams = pd.read_excel('C:\\GitHubRepos\\npi\\Analysis\\Controlled Environment\\CampVrnParams.xlsx',engine='openpyxl',index_col=0)    
# # Brown etal tests
# ## Bring in final leaf number and Vrn gene expression data
BaseHSTT = [0,2.5,7,14,18]
BaseTT = [110,335,884,2500,3430]
PhenoStages = pd.read_csv('C:\\GitHubRepos\\npi\\Analysis\\Controlled Environment\\PhenoStages.csv')
FLNData = PhenoStages.loc[:,['PhotoP','Vern','Cult','FLN']]
FLNData.set_index(['Cult','Vern','PhotoP'],inplace=True)
Cultivars = list(set(FLNData.index.get_level_values(0)))
PhenoStages.loc[:,'FLTT'] = [np.interp(PhenoStages.loc[x,'FLN'],BaseHSTT,BaseTT)for x in PhenoStages.index]
PhenoStages.loc[:,'TSTT'] = [np.interp(PhenoStages.loc[x,'HSTS'],BaseHSTT,BaseTT)for x in PhenoStages.index]
PhenoStages.loc[:,'FITT'] = [np.interp(PhenoStages.loc[x,'HSFI'],BaseHSTT,BaseTT)for x in PhenoStages.index]
PhenoStages.set_index(['PhotoP','Vern','Cult'],inplace=True)
VrnExpression = pd.read_pickle('C:\\GitHubRepos\\npi\\Analysis\\Controlled Environment\\VrnExpression.pkl')
VrnExpression.columns.values[0]='Vrn1'
VrnExpression.columns.values[1]='Vrn2'
VrnExpression.columns.values[2]='Vrn3'
Pps = VrnExpression.Pp.drop_duplicates()
Verns = VrnExpression.Vern.drop_duplicates()
# ## Make a dataframe of model outputs for each treatment
Cultivars
```

```python
def RunSingleScenario(D1,T1,P1,T2,P2,CultName,CultParams):
    Days = 200
    Day = range(Days)
    Tt1 = [T1] * D1; Pp1 = [P1] * D1
    Tt2 = [T2] * (Days - D1); Pp2 = [P2] * (Days - D1)
    Tt = Tt1 + Tt2; Pp = Pp1 + Pp2
    outputs = camp.CAMPmodel('',Day,Tt,Pp,CultParams,camp.CampConstants,90,0.000153173)
    outputs.loc[:,'Cultivar'] = CultName
    Vname = lambda V1: 'Full' if (V1==1) else 'Nil'
    outputs.loc[:,'Vernalisation'] = Vname(T1)
    outputs.loc[:,'Photoperiod'] = P2
    outputs.set_index(['Cultivar','Vernalisation','Photoperiod'],inplace=True)
    return outputs

BrownPred = pd.DataFrame(columns = ['Cultivar','Vernalisation','PhotoPeriod'])
BrownPred.set_index(['Cultivar','Vernalisation','PhotoPeriod'],inplace=True)

def CreateCultivarset(CultName,CultPars):
    FVLP = RunSingleScenario(90,1,0,23.6,16,CultName,CultPars)
    FVSP = RunSingleScenario(90,1,0,23.6,8,CultName,CultPars)
    NVLP = RunSingleScenario(90,23.6,16,23.6,16,CultName,CultPars)
    NVSP = RunSingleScenario(90,23.6,8,23.6,8,CultName,CultPars)
    DataTable = pd.concat([FVLP,FVSP])
    DataTable = pd.concat([DataTable,NVLP])
    DataTable = pd.concat([DataTable,NVSP])
    return DataTable

for C in Cultivars:
    BrownPred = pd.concat([BrownPred, CreateCultivarset(C, CultParams.loc[C,:])])
```
