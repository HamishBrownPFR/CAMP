```python
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math 
import MathsUtilities as MUte
import CAMP as camp
%matplotlib inline
pd.set_option('display.max_rows',1000)
```

```python
CampVrnParams = pd.read_excel('C:\\GitHubRepos\\npi\\Analysis\\Controlled Environment\\CampVrnParams.xlsx',index_col=0) 
CampVrnParams.sort_index(inplace=True)
```

```python
def RunSingleScenario(D1,T1,P1,T2,P2,CultName,CultParams,LightIntensity):
    Days = 300
    Day = range(Days)
    Tt1 = [T1] * D1; Pp1 = [P1] * D1
    Tt2 = [T2] * (Days - D1); Pp2 = [P2] * (Days - D1)
    Tt = Tt1 + Tt2; Pp = Pp1 + Pp2
    outputs = camp.CAMPmodel('FLN',Day,Tt,Pp,CultParams,camp.CampConstants,90,LightIntensity)
    return outputs
```

# LaTrobe tests

```python
LaTrobeCultivars = CampVrnParams.loc[CampVrnParams.Expt=='LaTrobe'].index.values
```

## Make a dataframe of model outputs for each treatment

```python
LaTrobePred = pd.DataFrame(columns = ['Cultivar','Vernalisation','PhotoPeriod'])
LaTrobePred.set_index(['Cultivar','Vernalisation','PhotoPeriod'],inplace=True)

for C in LaTrobeCultivars:
    print(C)
    CultPars = CampVrnParams.loc[C,:]
    LaTrobePred.loc[(C,'V','L'),'PredFLN'] = RunSingleScenario(60,5.8,16,22,17,C,CultPars,0.00006564)
    LaTrobePred.loc[(C,'V','S'),'PredFLN'] = RunSingleScenario(60,5.8,8,22,8,C,CultPars,0.00006564)
    LaTrobePred.loc[(C,'N','L'),'PredFLN'] = RunSingleScenario(200,22,16,22,17,C,CultPars,0.00006564)
    LaTrobePred.loc[(C,'N','S'),'PredFLN'] = RunSingleScenario(200,22,8,22,8,C,CultPars,0.00006564)
```

```python
colors = ['r','b','r','b']
fills = ['r','b','w','w']
lines = ['--','--','-','-']
Graph = plt.figure(figsize=(18,40))
pos=1
for cul in LaTrobeCultivars:
    ax = Graph.add_subplot(13,6,pos)
    pos+=1
    plt.plot(CampVrnParams.loc[cul,'FLN_LV'], LaTrobePred.loc[(cul,'V','L'),'PredFLN'],'o',mec='b',mfc='b',ms=5)
    plt.plot(CampVrnParams.loc[cul,'FLN_SV'], LaTrobePred.loc[(cul,'V','S'),'PredFLN'],'o',mec='b',mfc='w',ms=5)
    plt.plot(CampVrnParams.loc[cul,'FLN_LN'], LaTrobePred.loc[(cul,'N','L'),'PredFLN'],'o',mec='r',mfc='r',ms=5)
    plt.plot(CampVrnParams.loc[cul,'FLN_SN'], LaTrobePred.loc[(cul,'N','S'),'PredFLN'],'o',mec='r',mfc='w',ms=5)
    plt.plot([5,20],[5,20],'-',color='k')
    plt.text(5,18,cul)

Graph.patch.set_facecolor('white')
```

```python
for t in LaTrobePred.index:
    LaTrobePred.loc[t,'ObsFLN'] = CampVrnParams.loc[t[0],'FLN_'+t[2]+t[1]]
LaTrobePred.loc[:,'Residual'] = LaTrobePred.loc[:,'PredFLN'] - LaTrobePred.loc[:,'ObsFLN']

Graph = plt.figure(figsize=(18,40))
pos=1
for p in CampVrnParams.columns[:-1]:
    ax = Graph.add_subplot(11,4,pos)
    pos+=1
    xmin=100
    xmax=-100
    for c in CampVrnParams.loc[CampVrnParams.Expt=='LaTrobe'].index:
        xmin = min(xmin,CampVrnParams.loc[c,p])
        xmax = max(xmax,CampVrnParams.loc[c,p])
        plt.plot(CampVrnParams.loc[c,p], LaTrobePred.loc[(c,'V','L'),'Residual'],'o',mec='b',mfc='b',ms=5)
        plt.plot(CampVrnParams.loc[c,p], LaTrobePred.loc[(c,'N','L'),'Residual'],'o',mec='r',mfc='r',ms=5)
        plt.plot(CampVrnParams.loc[c,p], LaTrobePred.loc[(c,'V','S'),'Residual'],'^',mec='b',mfc='w',ms=5)
        plt.plot(CampVrnParams.loc[c,p], LaTrobePred.loc[(c,'N','S'),'Residual'],'^',mec='r',mfc='w',ms=5)
    plt.plot([xmin,xmax],[0,0],'-',color='k')
    plt.title(p)

Graph.patch.set_facecolor('white')
```
